```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the Windkessel model
R1 = 0.1            # Resistance of the large arteries (mmHg/(mL/s))
R2 = 0.2            # Resistance of the small arteries (mmHg/(mL/s))
C_hiper = 0.001     # Compliance of the arteries (mL/mmHg) - hypertension condition
C_normal = 0.005    # Compliance of the arteries (mL/mmHg) - normal condition
C_hipo = 0.01       # Compliance of the arteries (mL/mmHg) - hypotension condition
timestep = 0.001    # Time step (s)
total_time = 10     # Total simulation time (s)
Q_in = 0.1          # Inflow rate (mL/s)

# Arrays to store results
time = np.arange(0, total_time, timestep)

# Simulate and plot the Windkessel model under different conditions

# Hypertension
P0_hypertension = 150  # Initial arterial pressure for hypertension (mmHg)
pressure_hypertension = np.zeros(len(time))
flow_hypertension = np.zeros(len(time))

# Normal
P0_normal = 110  # Initial arterial pressure for normal (mmHg)
pressure_normal = np.zeros(len(time))
flow_normal = np.zeros(len(time))

# Hypotension
P0_hypotension = 60  # Initial arterial pressure for hypotension (mmHg)
pressure_hypotension = np.zeros(len(time))
flow_hypotension = np.zeros(len(time))

for i in range(1, len(time)):
# dP is the change in arterial pressure (in mmHg)
# dQ is the change in blood flow (in mL/s)

    # Hypertension
    dP_hypertension = (Q_in - (pressure_hypertension[i - 1] - P0_hypertension) / R1 - (pressure_hypertension[i - 1] - P0_hypertension) / C_hiper) * timestep
    dQ_hypertension = ((pressure_hypertension[i - 1] - P0_hypertension) / C_hiper - (pressure_hypertension[i - 1] - P0_hypertension) / R2) * timestep
    pressure_hypertension[i] = pressure_hypertension[i - 1] + dP_hypertension + dQ_hypertension
    flow_hypertension[i] = flow_hypertension[i - 1] + dQ_hypertension

    # Normal
    dP_normal = (Q_in - (pressure_normal[i - 1] - P0_normal) / R1 - (pressure_normal[i - 1] - P0_normal) / C_normal) * timestep
    dQ_normal = ((pressure_normal[i - 1] - P0_normal) / C_normal - (pressure_normal[i - 1] - P0_normal) / R2) * timestep
    pressure_normal[i] = pressure_normal[i - 1] + dP_normal + dQ_normal
    flow_normal[i] = flow_normal[i - 1] + dQ_normal

    # Hypotension
    dP_hypotension = (Q_in - (pressure_hypotension[i - 1] - P0_hypotension) / R1 - (pressure_hypotension[i - 1] - P0_hypotension) / C_hipo) * timestep
    dQ_hypotension = ((pressure_hypotension[i - 1] - P0_hypotension) / C_hipo - (pressure_hypotension[i - 1] - P0_hypotension) / R2) * timestep
    pressure_hypotension[i] = pressure_hypotension[i - 1] + dP_hypotension + dQ_hypotension
    flow_hypotension[i] = flow_hypotension[i - 1] + dQ_hypotension

# Plot the results
plt.figure(figsize=(12, 6))

plt.subplot(3, 2, 1)
plt.plot(time, pressure_hypertension)
plt.title('Arterial Pressure (Hypertension)')
plt.xlabel('Time (s)')
plt.ylabel('Pressure (mmHg)')

plt.subplot(3, 2, 2)
plt.plot(time, flow_hypertension)
plt.title('Blood Flow (Hypertension)')
plt.xlabel('Time (s)')
plt.ylabel('Flow (mL/s)')

plt.subplot(3, 2, 3)
plt.plot(time, pressure_normal)
plt.title('Arterial Pressure (Normal)')
plt.xlabel('Time (s)')
plt.ylabel('Pressure (mmHg)')

plt.subplot(3, 2, 4)
plt.plot(time, flow_normal)
plt.title('Blood Flow (Normal)')
plt.xlabel('Time (s)')
plt.ylabel('Flow (mL/s)')

plt.subplot(3, 2, 5)
plt.plot(time, pressure_hypotension)
plt.title('Arterial Pressure (Hypotension)')
plt.xlabel('Time (s)')
plt.ylabel('Pressure (mmHg)')

plt.subplot(3, 2, 6)
plt.plot(time, flow_hypotension)
plt.title('Blood Flow (Hypotension)')
plt.xlabel('Time (s)')
plt.ylabel('Flow (mL/s)')

plt.tight_layout()
plt.show()
```
